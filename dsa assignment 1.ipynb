### Load the data from the CSV file using NumPy

```python
data = np.loadtxt('iris.csv', delimiter=',', skiprows=1, usecols=(0, 1, 2, 3))
```

```python
# Print the first 5 rows of the data
print('Data:')
print(data[:150])
```

```python
# Get the mean, median, and standard deviation of each column
means = np.mean(data, axis=0)
medians = np.median(data, axis=0)
stds = np.std(data, axis=0)
```

```python
# Print the results
print('\nMeans:')
print(means)
print('\nMedians:')
print(medians)
print('\nStandard Deviations:')
print(stds)
```
